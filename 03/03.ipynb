```python
import pandas as pd
```

```python
df = pd.read_csv('rucksacks.csv')
```

```python
def findd(row):
    items = row[0]
    half = len(items)//2
    first, second = items[:half], items[half:]
    for i in range(half):
        if first[i] in second:
            return first[i]
```

```python
df['error'] = df.apply(findd, axis=1)
```
